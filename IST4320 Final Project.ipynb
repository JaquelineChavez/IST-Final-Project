```python
import tkinter as tk
from tkinter import ttk
main = tk.Tk()
main.geometry('500x200')
label = tk.Label(main, text="Calculate your grade!")


Earned_points = tk.Label(main, text="To get started, please enter your total points earned!",
                   font=('helvetica',15, 'bold'))

Earned_var = tk.IntVar()
Total_var = tk.IntVar()

Earned_entry = tk.Entry(main,
                    textvariable = Earned_var,
                    font=('helvetica',12, 'normal'))

Total_points = tk.Label(main, text="Next, please enter the total points possible in the class!",
                     font=('helvetica',15, 'bold'))

Total_entry = tk.Entry(main,
                    textvariable = Total_var,
                    font=('helvetica',12, 'normal'))
Entry_result = tk.Entry(main,
                       font=('helvetica',15, 'normal'))
button1 = tk.Button(main, 
                    text='=',
                   font=('helvetica',10,'bold'))
def firstCalculation():
    Ans = int(Earned_entry.get()) / int(Total_entry.get())
    
    Entry_result.insert(0, Ans)
copy_var = tk.IntVar()
percentage_var = tk.IntVar()

copy_result = tk.Label(main, 
                       text="Next, copy the first result but only using up to the fourth digit after the period (Ex: 0.1234)", 
                       font=('helvetica',15, 'bold'))
copy_entry = tk.Entry(main, 
                      textvariable = copy_var, 
                      font=('helvetica',15, 'normal'))
percentage_result = tk.Label(main, 
                       text="Next, enter '100'", 
                       font=('helvetica',15, 'bold'))
percentage_entry = tk.Entry(main, 
                      textvariable = percentage_var, 
                      font=('helvetica',15, 'normal'))
Final_result = tk.Entry(main,
                       font=('helvetica',15, 'normal'))
button2 = tk.Button(main, 
                    text='=',
                   font=('helvetica',10,'bold'))
def SecondCalculation():
    Ans = float(copy_entry.get()) * int(percentage_entry.get())
    
    Final_result.insert(0, Ans)
import sqlite3
sql_connect = sqlite3.connect('database.db')
cur = sql_connect.cursor() 

cur.execute('''CREATE TABLE IF NOT EXISTS projectapp_data (
first_calc FLOAT,
percentage FLOAT
)
''')
sql_connect.commit()
def getInfo():
    first_calc = Entry_result.get()
    percentage = Final_result.get()
    
    cur.execute("INSERT INTO projectapp_data (first_calc,percentage) VALUES (?,?)",
            (first_calc,percentage))
    sql_connect.commit()

    Entry_result.set("")
    Final_result.set("")
    main.destroy()
submit_button = tk.Button(main,
                          text="Submit",
                          font=('helvetica',10, 'normal'),
                          command = getInfo)


button1['command'] = firstCalculation
button2['command'] = SecondCalculation
label.pack()
Earned_points.pack()
Earned_entry.pack()
Total_points.pack()
Total_entry.pack()
Entry_result.pack()
button1.pack()
copy_result.pack()
copy_entry.pack()
percentage_result.pack()
percentage_entry.pack()
Final_result.pack()
button2.pack()
submit_button.pack()
main.mainloop()
```
